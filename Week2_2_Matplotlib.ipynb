When plotting, if the data exceeds the memory size we run into a problem. To overcome this, we use sampling.


Sampling is taking the subset of the original data. Due to inherent randomness, preserves most properties of original data. 

Reduces Computational cost



```python
arr = [5,5,5,5,6,7,8,2,5,6,7,8,4,5,6,7,8,2,3,4,5,6,7,6,5,4,1,2,3,1,3,4,]

#To paste it directly under this code box
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
plt.boxplot(arr)
plt.show()

#Mean is the central line
#The box is the standard deviation where half of all data lies in
```
